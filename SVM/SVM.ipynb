```python
def price_volatility(number_of_days, closing_price_array):
    volatility = []
    moving_volatility = []
    
    for i in range (1, number_of_days+1):
        percent_change = 100 * ((closing_price_array[i] - closing_price_array[i - 1]) / closing_price_array[i -1])
        moving_volatility.append(percent_change)
    volatility.append(np.mean(moving_volatility))
    
    for i in range (number_of_days+1, len(closing_price_array)-1):
        del moving_volatility[0]
        percent_change = 100 * ((closing_price_array[i] - closing_price_array[i - 1]) / closing_price_array[i -1])
        moving_volatility.append(percent_change)
        volatility.append(np.mean(moving_volatility))
        
    return volatility, moving_volatility
```
